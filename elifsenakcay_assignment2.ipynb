```r
library(tidyverse)
library(magrittr)
library(DT)
library(data.table)
tswift<-fread("http://raw.githubusercontent.com/rfordatascience/tidytuesday/refs/heads/master/data/2023/2023-10-17/taylor_all_songs.csv")
tswift
```

Make a select operation of columns, I selected 3 columns, "liveness","valence","tempo"

Make a filter operation of rows, I filtered the rows to show the owner artists of the song who are not Taylor.

```r
tswift %>% select("liveness","valence","tempo")
```

```r
tswift %>% filter(artist != "Taylor Swift")
```

Make a mutate operation to create a new column, I created a new column and defined it as tempo/valence.




```r
tswift %<>% mutate(t_v_rate= tempo/valence)
tswift
```

Make a group_by and summarize operation, I grouped the data according to the key of the songs, major or minor and then find the average of the values.



```r
tswift %>% group_by(mode_name)%>% summarise(mean(tempo, na.rm=TRUE)) 
```

Make a pivot_longer operation, I keep the track name and make the data wider by using the key modes.

Make a pivot_wider operation,


```r
tswift2<- tswift %>% pivot_wider(id_cols = track_name , names_from = key_mode, values_from = duration_ms)
tswift2
```

```r
tswift2 %>% pivot_longer(cols =-"track_name" , names_to = "key_note2", values_to = "duration_ms2", values_drop_na = T)
```

Make a left_join operation with another table with common columns (you can create a new table out of the original one for this purpose, if you don't have a second table)

```r
album <- fread("http://raw.githubusercontent.com/rfordatascience/tidytuesday/refs/heads/master/data/2023/2023-10-17/taylor_albums.csv")
album 
tswift %>% left_join(album, by = "album_name")
```
